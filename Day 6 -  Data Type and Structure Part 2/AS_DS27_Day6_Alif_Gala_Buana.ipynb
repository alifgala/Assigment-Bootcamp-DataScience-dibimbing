## Soal Nomor 1

```python
# Barplot untuk Gender (Nominal) slicing data (10)
sns.countplot(x='City', data=df, order = df['City'].value_counts().index[:10], hue= 'City', palette= sns.color_palette())
plt.title('Barplot of City')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

## insight
#### data yang ditampilkan merupakan 10 urutan row nama kota/city yang memiliki data terbanyak di dataframe, dari dataframe tersebut terlihat pada orang-orang yang terdata tersebut paling banyak berasal dari Banda Aceh yaitu sebanyak 546 orang dan paling sedikit berasal dari makasar yaitu 505 orang, akan tetapi dapat kita lihat penyebaran lokasi kota dari dataframe  relatif merata dan tidak berselisih jauh diantara 10 kota tersebut.

##  While loop

```python
i = 0
while i <len (df['Age']): #fungsi while loop dengan parameter len.
    i+=1
    if i <100:
        print (df['Age'].iloc[i], df['Marital Status'].iloc [i])
```

## For loop

```python
for index, row in df.iloc[:100].iterrows(): #fungsi for loop untuk dataframe dengan indeks 0-100
    if row['Age'] <= 50:
        print(row['Age'], row['Marital Status'])
```

## Nested Loop

```python
for x_index, x_row in df.iloc[0:100].iterrows(): #looping pertama
    L1 = x_row['Salary']
    L2 = x_row['Marital Status']
    for y_index, y_row in df.iloc[x_index+1:100].iterrows(): # Loooping kedua
        L3 = y_row['Salary']
        L4 = y_row['Marital Status']
        print (L3,L4)
```
